#### Montando vários gráficos em ume mesma janela

```python
# Definindo o Intervalo das Funções
np.arange(-10,10)

# Definindo as novas funções

y= 2*x

z=x**5+7

w=2*x**2-1

#Montando a janela de gráficos
# - Os gráfics são montados em uma única figura , como se fosse uma matriz de n linhas x 1 coluna
# a sintaxe do comando e plt.subplot(< número de gráficos>,1) , no caso o número de gráficos a serem plotados e 1 representa 1 coluna
# no caso de três gráficos na mesma janela temos plt.subplot(31"x"), onde 3 é o número de linhas, 1 e uma coluna, e "X"
# é o número do gráfico a ser plotado

plt.subplot(311),plt.title("Mais Gráficos"), plt.grid(),plt.plot(x,y)
plt.subplot(312),plt.grid(),plt.plot(x,z)
plt.subplot(313),plt.grid(),plt.plot(x,w)
# salvando os novos gráficos
plt.savefig("Gráficos2.png")
```

#### Aplicações do Numpy : Calculando o retorno diário de um grupos de ativos financeiros

```python
# Conjunto de retornos diários de ativos financeiros
resultado_diario= [10,2,-4,5,1,15, 7] # lista com o resultado de 7 dias de movimento de um ativo financeiro

# gerando um array destes resultados
vetor =np.array(resultado_diario)

# calculanado o retorno dos ativos financeiros
retorno_diario = vetor[1:7]-vetor[0:6]/vetor[0:6]
print(retorno_diario)
```

#### Gráfico do Retorno Dário


```python
plt.plot(retorno_diario),plt.title("Retorno Dário"),plt.grid(), plt.savefig("fig4.png")
```
